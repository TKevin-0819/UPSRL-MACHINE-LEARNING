```python
#Importaciones de procesamientos de dadtos e imagens
import numpy as np
import matplotlib.pyplot as plt

# Importacion de Scikit_Learn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree  import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.tree import plot_tree
```

## The Iris dataset consists of 150 samples of iris flowers from three different species: Setosa, Versicolor, and Virginica. Each sample includes four features: sepal length, sepal width, petal length, and petal width. It was introduced by the British biologist and statistician Ronald Fisher in 1936 as an example of discriminant analysis.

```python
# Descarga de nuestras muestras

data = load_iris()
```

```python
print(data)
```

```python
#Creacion de variables para el modelo
x = data.data
y = data.target
feature_names = data.feature_names
target_names = data.target_names
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    random_state=42
    )
```

```python
model = DecisionTreeClassifier(
    criterion='gini',
    max_depth=3,
    random_state=42
)

model.fit(x_train, y_train)
```

```python
# generacion de preiccion
y_pred = model.predict(x_test)
```

```python
#evaluacion del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"accuracy:{accuracy}")
print(classification_report(y_test, y_pred))
```

```python
plot_tree(model)
```
