```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# 1. tenemos una base de datos con 500 muestras
# 2. tenemos dos features
# 3. no hay features redundantes
# 4. numero de features informativos
X, y = make_classification(
    n_samples=500,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    random_state=42
)
print(X)
print(y)
```

```python
# Dividimos datos para entrenamiento y pruebas
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)
print(X_train)
print(X_test)
```

```python
# Escalamos datos para procesar (preprocesamiento)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
print(X_train)
print(X_test)
```

```python
# Generamos nuestro modelo y lo alimentamos
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Generamos prediccion
y_pred = model.predict(X_test)
print(y_pred)

```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:", confusion_matrix(y_test, y_pred))
print("Classification report:", classification_report(y_test, y_pred))



```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# 1. Generamos la base de datos
X, y = make_classification(
    n_samples=500,
    n_features=2,
    n_redundant=0,
    n_informative=2,
    n_clusters_per_class=1,
    random_state=42
)

print(X)
print(y)
```

```python
# 2. Dividimos datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42
)

print(X_train)
print(X_test)
```

```python
# 3. Creamos el pipeline (preprocesamiento + modelo)
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression())
])
```

```python
# 4. Entrenamos el modelo
pipeline.fit(X_train, y_train)

# 5. Prediccin
y_pred = pipeline.predict(X_test)
print(y_pred)
```

```python
# 6. Mtricas
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification report:\n", classification_report(y_test, y_pred))
```

```python
# 7. Funcin para graficar la frontera de decisin (adaptada al pipeline)
def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx, yy = np.meshgrid(
        np.linspace(x_min, x_max, 200),
        np.linspace(y_min, y_max, 200)
    )

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors="k", s=50)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("Logistic Regression Decision Boundary (Pipeline)")
    plt.show()
```

```python
# 8. Graficamos usando el pipeline completo
plot_decision_boundary(X, y, pipeline)
```
