Importacion de librerias

Este script importa las librerías necesarias para el análisis.
numpy se usa para manejar arreglos y datos numéricos.
matplotlib.pyplot sirve para graficar los datos y resultados.
LinearRegression permite crear y entrenar el modelo lineal.
mean_squared_error y r2_score se usan para evaluar el desempeño del modelo.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

El script define los datos para una regresión lineal.
x es la variable independiente, se convierte en un arreglo 2D porque Scikit-Learn lo requiere.
y es la variable dependiente, con los valores reales que el modelo debe aprender.
Los print solo sirven para verificar los datos antes de entrenar el modelo.

```python
# y = m * x + b

# Variable independiente
# SciKit-Learn requiere que se usen arreglos de 2D
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
print(x)

# Variable dependiente
y = np.array([2, 4, 5, 4, 5])
print(y)
```

Este script grafica los datos originales.
plt.scatter(x, y) dibuja los puntos reales en el plano.
xlabel y ylabel nombran los ejes.
title pone el título de la gráfica (solo tiene un error ortográfico: originales).
grid(True) agrega la cuadrícula para facilitar la lectura.
show() muestra la gráfica en pantalla.

```python
plt.scatter(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Datos origuinales")
plt.grid(visible=True)
plt.show()
```

Este script crea y entrena un modelo de regresión lineal.
LinearRegression() inicializa el modelo.
model.fit(x, y) entrena el modelo con los datos.
model.coef_[0] obtiene la pendiente (m) de la recta.
model.intercept_ obtiene el intercepto (b).
Los print muestran los valores de m y b con dos decimales.

```python
# Generamos nuestro modelo de regrecion linela
model = LinearRegression()

# Alimentamos nuestro modelo
model.fit(x, y)

# Parametros:m(peso), b(sesgo)
m = model.coef_[0]
b = model.intercept_


print(f"Pendiente (m): {m:.2f}")
print(f"intercepto (b): {b:.2f}")
```

Este script genera las predicciones del modelo y evalúa su desempeño.
model.predict(x) calcula los valores predichos (y_pred).
mean_squared_error mide el error promedio al cuadrado entre valores reales y predichos.
r2_score indica qué tan bien el modelo explica los datos (entre 0 y 1).
Los print muestran ambas métricas con dos decimales.

```python
# Generemaos prediccion
y_pred = model.predict(x)
# Aliemntamos esa prediccion
nse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error: {nse:.2f}")
print(f"R^2 Score: {r2:.2f}")
```

```python
Este script muestra la comparación entre los datos reales y el modelo lineal.
Los puntos (scatter) representan los datos reales.
La línea roja (plot) muestra las predicciones del modelo.
Se etiquetan los ejes, se agrega título, cuadrícula y leyenda para mostrar qué representa cada elemento.
```

```python
plt.scatter(x, y, label="Datos reales")
plt.plot(x, y_pred, color="red", label="Modelo lineal")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Datos originales")
plt.grid(visible=True)
plt.legend()
plt.show()
```
