En este bloque medimos el tiempo de ejecución de una suma de una lista de Python estándar.Iniciamos creando una lista de 100 millones de números, pra despues utiliza time.perf_counter() para medir el tiempo que tarda un bucle for en recorrer la lista y sumar sus elementos (más 1 extra en cada iteración). POr ultimo imprime el tiempo total transcurrido.

```python
lista = list(range(100_000_000))

inicio = time.perf_counter()

suma = 0
for i in lista:
    suma +=i
    suma +=1
final1 = time.perf_counter()

tiempo = final1 - inicio
print(f"tiempo de ejecucion: {tiempo:.6f} sec")
```

Este bloque compara el rendimiento de la suma con numpy. Crea un array de NumPy (np_lista) de 100 millones de números, despues mide el tiempo que tarda la función optimizada np.sum() de NumPy en calcular la suma de todos los elementos y muestra la comparacion en eficiencia de NumPy frente a los bucles de Python nativos.

```python
np_lista = np.arange(100_000_000)

inicio = time.perf_counter()

suma_np = np.sum(np_lista)

final = time.perf_counter()

tiempo = final1 - inicio
print(f"tiempo de ejecucion: {tiempo:.6f} sec")
```

En este blque creamos una base de datos. Empezamos creando una lista llamada base_de_datos que contiene información sobre mis compañeros.(nombre, edad, mascota).despues definimos dos columnas y filas para acceder mas rapidoa los nombres de columnas y filas de sus respectivos índices, de esta manera facilitando el acceso a los datos de la lista de forma legible. al final se  imprime  'Kevin'

```python
base_de_datos = [
    ["Oscar", 21, "Gatos"],
    ["Paula", 21, "Tiburones"],
    ["Jonatan", 24, "Hurones"],
    ["Kevin", 22, "Periquitos"],
    ["Lucero", 21, "Perros"],
    ["Israel", 21, "Lobos"],
    ["Sergio", 21, "Perros"],
    ["Yahir", 21, "Leones"],
    ["Diego", 21, "Pingüinos"]
]
columnas = {"Nombre":0, "Edad":1, "Mascota":2}
filas = {"Oscar":0,"Paula":1,"Jonatan":2,"Kevin":3,"Lucero":4,"Israel":5,"Sergio":6,"Yahir":7}
print(base_de_datos[filas["Kevin"]][columnas["Mascota"]])
```

Este bloque convierte la base_de_datos en un DataFrame de pandas para un manejo de datos más eficiente y estructurado. Crea un dF (DataFrame) usando la lista base_de_datos, asigna nombres a las columnas ("Nombre", "Edad", "Animal") y un índice del 1 al 10, despues imprime el DataFrame completo, se muestra cómo filtrar el DataFrame para obtener la mascota de 'Kevin' utilizando la sintaxis de pandas.



```python
dF = pd.DataFrame(base_de_datos, columns=["Nombre", "Edad", "Animal"], index=range(1,10))
print(dF)
print(dF.Animal[dF['Nombre'] == 'Kevin'])
```
